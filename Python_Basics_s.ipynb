Question 2: Explain the difference between List and Tuple in Python.

List: You can change a list after creating it (add, remove, or edit items). It uses square brackets [].

Tuple: You cannot change a tuple once it is made. It is "fixed." It uses parentheses ().


Question 3: What is a function in Python? Why are functions useful?
A function is a block of code that only runs when you call it.

Why they are useful: They help you avoid writing the same code over and over. You write the logic once and reuse it whenever you need it.

Question 5: Program to check if a number is even or odd.

```python
num = int(input("Enter a number: "))
if num % 2 == 0:
    print("This is an Even number.")
else:
    print("This is an Odd number.")
```

Question 10: Count how many times a number appears in a list.

```python
example_list = [2, 3, 4, 2, 5, 2]
target = 2
count = example_list.count(target)
print("The number", target, "appears", count, "times.")
```
